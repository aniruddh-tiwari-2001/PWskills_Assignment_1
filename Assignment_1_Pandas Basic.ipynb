### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

**DataFrame** :- A data frame in pandas is a 2 dimensional labled data structure which is similar to tables or spredsheets. It consist of rows and columns, whare each column can contain different data types. 

**Differnce Between DataFrame and Series:**

- Data structure: A DataFrame is a 2-dimensional data structure, while a Series is a 1-dimensional data structure. This means that a DataFrame has rows and columns, while a Series only has columns.

- Index: A DataFrame has an index, which is a unique identifier for each row. A Series does not have an index by default, but you can add one if you want.

- Data types: A DataFrame can contain multiple data types, while a Series can only contain a single data type.

- Methods: A DataFrame has a wider range of methods than a Series. This is because DataFrames are more complex data structures and require more functionality.


```python
## Creating a data frame:
data = {'Name':['Alice','Bob','Claire'], 'Age':[25,30,27], 'Gender':['Female','Male','Female']}
df = pd.DataFrame(data)
print("This is Data Frame")
print(df)
print("\n")
## Creating a Series:
print("This is Series")
df[['Age']]
```

### Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
# Creating a Series of fisrt name, last name, age, education, gender
# Series is created by pd.Series() function
s_name = pd.Series(['Aniruddh', 'Alish', 'Aniket', 'Ayush', 'Anuj', 'Rishi'])
s_lname = pd.Series(['Tiwari', 'Tiwari', 'Tiwari', 'Pandey', 'Tiwari', 'Pandey'])
s_age = pd.Series([22,20,18,17,12,15])
s_edu = pd.Series(['B.tech','BCA','12th','12th','6th','8th'])
s_gender = pd.Series(['Male','Female','Male','Male','Male','Male'])

# Store the Series data into a dictionary
data = {'First Name':s_name, 'Last Name':s_lname, 'Age':s_age, 'Education':s_edu, 'Gender':s_gender}
# Used pd.DataFrame() function to create the DataFrame
df = pd.DataFrame(data)
print(df)
```
